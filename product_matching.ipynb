В рамках проекта предстоит поработать с реальными сырыми данными от одного из крупнейших маркетплейсов страны.

Нас ждет интересная задача сопоставления и поиска наиболее похожих товаров.
Сопоставление или “мэтчинг” (англ. matching - соответствия) - одна из базовых задач машинного обучения, которая встречается в информационном поиске, компьютерном зрении, рекомендательных системах и др.

Будем работать с алгоритмами приближённого поиска ближайших соседей, создавать индексы в векторных базах данных и обучать ранжирующие модели. 

Дано: 
Два множества объектов: A и B. Каждый объект в множествах описывается какими-то признаками.

Желаемый результат: 
Для каждого объекта из множества A найти один или несколько объектов из B, которые близки к нему по заданной метрике.

base.csv - анонимизированный набор товаров. Каждый товар представлен как уникальный id (0-base, 1-base, 2-base) и вектор признаков размерностью 72.

validation.csv - датасет с товарами (уникальный id и вектор признаков), для которых надо найти наиболее близкие товары из base.csv

validation_answer.csv - правильные ответы к предыдущему файлу.


Конкретизируем задачу:  

- необходимо разработать алгоритм, который для всех товаров из validation.csv предложит несколько вариантов наиболее похожих товаров из base;

- а также оценить качество алгоритма по метрике accuracy@5


Метрика: 

$$
accuracy@5 = 100 * \frac{\text{кол-во верно определённых похожих объектов из 5 возможных}} {5}
$$

**Перейдем к более детальному исследовательскому анализу:**

- Т.к. датафреймы `df_train` и `df_val` являются подвыборками `df_base`, ограничимся выводом графиков для этого датафрейма.
- Т.к. `df_base` содержит большое кол-во объектов, для построения гистограмм и диаграмм размаха найдём процент выборки, при котором подвыборка будет репрезентативна.

Выполним тест Колмогорова-Смирнова и Z-тест для каждого размера подвыборки и определяет, начиная с какого процента выборки эти тесты становятся статистически значимыми (p-value меньше заданного уровня значимости alpha).

```python
# Выбираем диапазон размеров подвыборок от 1% до 15%
min_percent = 10
max_percent = 15

# Определяем уровень значимости для тестов
alpha = 0.05

# Сохраняем результаты тестов для каждого размера подвыборки
ks_test_results = []
z_test_results = []

for percent in range(min_percent, max_percent + 1):
    # Разделяем данные на обучающую и тестовую выборки
    _, sample_data, _, _ = train_test_split(data_base, data_base, test_size=percent/100, random_state=42)
    
    # Выбираем только числовые столбцы
    numeric_columns = sample_data.select_dtypes(include=[np.number]).columns
    sample_data_numeric = sample_data[numeric_columns]
    
    # Преобразуем в одномерный массив
    sample_data_flat = sample_data_numeric.to_numpy().flatten()
    
    # Выполняем тест Колмогорова-Смирнова
    ks_statistic, ks_pvalue = kstest(sample_data_flat, 'norm')
    ks_test_results.append((percent, ks_statistic, ks_pvalue))
    
    # Выполняем Z-тест
    z_statistic = zscore(sample_data_flat)
    z_test_results.append((percent, z_statistic))

# Находим минимальный размер выборки, при котором результаты тестов становятся статистически значимыми
min_percent_significant_ks = None
min_percent_significant_z = None

for percent, ks_statistic, ks_pvalue in ks_test_results:
    if ks_pvalue < alpha:
        min_percent_significant_ks = percent
        break

for percent, z_statistic in z_test_results:
    z_pvalue = np.any(z_statistic < alpha)
    if z_pvalue:
        min_percent_significant_z = percent
        break

print(f"Минимальный процент выборки, при котором тест Колмогорова-Смирнова становится значимым: {min_percent_significant_ks}%")
print(f"Минимальный процент выборки, при котором Z-тест становится значимым: {min_percent_significant_z}%")
```

**Далее построим гистограммы, диаграммы размаха и вычислим процент выбросов для каждого численного признака в 10% случайной выборке из нашего исходного датафрейма.**

```python
# Получение случайной 10% выборки
sample_data = data_base.sample(frac=0.1, random_state=42)

# Выбор только численных признаков
numeric_features = sample_data.select_dtypes(include=[np.number])

# Построение гистограмм, диаграмм размаха и вычисление процента выбросов для каждого численного признака
for column in numeric_features.columns:
    plt.figure(figsize=(10, 6))
    
    # Гистограмма
    plt.subplot(1, 3, 1)
    sample_data[column].hist(bins=20, color='skyblue')
    plt.title(f'Гистограмма признака {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    
    # Диаграмма размаха
    plt.subplot(1, 3, 2)
    sample_data.boxplot(column=[column])
    plt.title(f'Диаграмма размаха признака {column}')
    
    # Вычисление процента выбросов
    Q1 = sample_data[column].quantile(0.25)
    Q3 = sample_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_percent = ((sample_data[column] < lower_bound) | (sample_data[column] > upper_bound)).mean() * 100
    print(f'Процент выбросов в признаке {column}: {outliers_percent:.2f}%')
    
    plt.show()
```

## Создание базовых моделей 

Для проведения тестов будем использовать базовые модели `FAISS` и `ANNOY`, т.к. они значительно превосходят связку `KMeans` + `NearestNeighbors` в скорости.
- Принцип работы `FAISS`:
  - Создание инвертированного индекса.
  - Поиск похожих сначала по центроидам, затем по выбранным кластерам.
- Принцип работы `ANNOY`:
  - Создание индекса из леса бинарных деревьев, состоящих из итеративно разделенённых равноудалёнными гиперплоскостями пространств.
  - Поиск похожих спуском по деревьям со сравнением узлов с искомым вектором.<br>

```python
# Функция вычисления целевой метрики accuracy@k.
def get_accuracy(df, df_eval, target,
                 model_type = None,
                 faiss_n_clust = None,
                 faiss_index = 'ivf_flat',
                 faiss_metric = 'euclidean',
                 faiss_train_sample_size = None,
                 faiss_pca_dims = None,
                 faiss_ef_search = 32,
                 annoy_distance_type = 'euclidean',
                 annoy_n_trees = 10,
                 annoy_search_n_nodes = -1,
                 k_similar = 5,
                 SEED = 42):


    # Изменим формат типа модели при необходимости.
    model_type = [model_type] if isinstance(model_type, str) else model_type

    # Определим длину вектора признаков.
    dims = df.shape[1]

    # Определим базовый индекс.
    base_index = {k: v for k, v in enumerate(df.index.to_list())}

    # FAISS
    if 'faiss' in model_type:
        # FAISS с PCA
        if faiss_pca_dims:
            # Снизим размерность данных и приведём данные в соответстие faiss.
            pca_matrix = faiss.PCAMatrix(dims, faiss_pca_dims)
            pca_matrix.train(df)
            assert pca_matrix.is_trained
            df_arr = pca_matrix.apply(df)
            df_eval_arr = pca_matrix.apply(df_eval)

            dims = faiss_pca_dims

        # FAISS с индексом IVFFLAT.
        if faiss_index == 'ivf_flat':
            # Определим квантизатор.
            quantizer = faiss.IndexFlatL2(dims)

            # FAISS с расстоянием Махаланобиса.
            if faiss_metric == 'mahalanobis':
                if faiss_pca_dims:
                    df = df_arr
                    df_eval = df_eval_arr
                # Вычислим матрицу ковариации.
                df_c = df - df.mean(0)
                cov = np.dot(df_c.T, df_c) / df_c.shape[0]

                # Преобразуем ковариационную матрицу.
                L = np.linalg.cholesky(cov)
                mahalanobis_transform = np.linalg.inv(L)

                # Функция преобразования данных.
                def inverse_cholesky_cov(df, mahalanobis_transform):
                    df_trans = np.dot(df, mahalanobis_transform.T)
                    return df_trans.astype('float32')

                # Преобразуем данные.
                df_arr = inverse_cholesky_cov(df,
                                              mahalanobis_transform)
                df_eval_arr = inverse_cholesky_cov(df_eval,
                                                   mahalanobis_transform)

                # Определим индекс.
                faiss_index = faiss.IndexIVFFlat(quantizer,
                                                 dims,
                                                 faiss_n_clust,
                                                 faiss.METRIC_L2)

            else:
                if not faiss_pca_dims:
                    # Приведём данные в соответстие faiss.
                    df_arr = np.ascontiguousarray(df.values).astype('float32')
                    df_eval_arr = (
                        np.ascontiguousarray(df_eval.values).astype('float32')
                        )

                # FAISS с косинусным расстоянием.
                if faiss_metric == 'cosine':

                    # Нормализуем векторы.
                    faiss.normalize_L2(df_arr)
                    faiss.normalize_L2(df_eval_arr)

                    # Определим индекс.
                    faiss_index = faiss.IndexIVFFlat(quantizer,
                                                    dims,
                                                    faiss_n_clust,
                                                    faiss.METRIC_INNER_PRODUCT)

                # FAISS с поиском максимального скалярного произведения.
                elif faiss_metric == 'mips':

                    # Определим индекс.
                    faiss_index = faiss.IndexIVFFlat(quantizer,
                                                    dims,
                                                    faiss_n_clust,
                                                    faiss.METRIC_INNER_PRODUCT)

                # FAISS с Евклидовым расстоянием.
                elif faiss_metric == 'euclidean':

                    # Определим индекс.
                    faiss_index = faiss.IndexIVFFlat(quantizer,
                                                    dims,
                                                    faiss_n_clust)
                    
                    
                    
                 # FAISS с метрикой Jaccard
                elif faiss_metric == 'jaccard':
                    # Приведём данные к типу np.uint8 для использования Jaccard
                    df_arr = np.ascontiguousarray(df.values).astype('uint8')
                    df_eval_arr = np.ascontiguousarray(df_eval.values).astype('uint8')

                    # Определим индекс
                    faiss_index = faiss.IndexIVFFlat(quantizer, dims, faiss_n_clust, faiss.METRIC_Jaccard)

                # FAISS с метрикой Bray-Curtis
                elif faiss_metric == 'braycurtis':
                    # Приведём данные к типу np.float32 для использования Bray-Curtis
                    df_arr = np.ascontiguousarray(df.values).astype('float32')
                    df_eval_arr = np.ascontiguousarray(df_eval.values).astype('float32')

                    # Определим индекс
                    faiss_index = faiss.IndexIVFFlat(quantizer, dims, faiss_n_clust, faiss.METRIC_BrayCurtis)


                else:
                    raise ValueError('Другие метрики "FAISS" не '+
                                     'поддерживаются.')

        # FAISS с индексом IVF_HNSW64.
        elif faiss_index == 'ivf_hnsw':
            # Приведём данные в соответстие faiss.
            df_arr = np.ascontiguousarray(df.values).astype('float32')
            df_eval_arr = (
                np.ascontiguousarray(df_eval.values).astype('float32')
                )

            # Определим индекс.
            faiss_index = faiss.index_factory(dims,
                                              (f'IVF{faiss_n_clust}_HNSW'+
                                               f'{faiss_ef_search}'+',Flat'))
            faiss_index.nprobe = 8

        start_faiss = perf_counter()

        faiss_index.train(df_arr[:faiss_train_sample_size, :])
        faiss_index.add(df_arr)

        # Построим индекс.
        vecs, idx = faiss_index.search(df_eval_arr, k_similar)

        # Вычислим целевую метрику accuracy@k для FAISS.
        acc_faiss = 0
        for t, el in zip(target.values.tolist(), idx):
            acc_faiss += int(t in [base_index[r] for r in el])

        # Вычислим время выполнения.
        time_faiss = perf_counter() - start_faiss

    # ANNOY
    if 'annoy' in model_type:

        # Определим индекс ANNOY.
        annoy_index = AnnoyIndex(dims, annoy_distance_type)
        annoy_index.set_seed(SEED)
        # Добавим векторы и их индексы.
        for i, v in zip(range(df.shape[0]), df.values):
            annoy_index.add_item(i, v)

        # Построим индекс.
        annoy_index.build(annoy_n_trees, n_jobs=-1)

        start_annoy = perf_counter()

        # Получим K похожих элементов для каждого запроса и собрём в список.
        idx = []
        for v in df_eval.values:
            idx.append(annoy_index.get_nns_by_vector(
                v,
                k_similar,
                search_k=annoy_search_n_nodes)
            )

        # Вычислим целевую метрику accuracy@k для ANNOY.
        acc_annoy = 0
        for t, el in zip(target.values.tolist(), idx):
            acc_annoy += int(t in [base_index[r] for r in el])

        # Вычислим время выполнения.
        time_annoy = perf_counter() - start_annoy

    # Выведем метрику и время исполнения для каждого/всех алгоритмов.
    if model_type == ['faiss']:
        return (round((100 * acc_faiss / len(idx)), 3),
                     round(time_faiss, 2))
    elif model_type == ['annoy']:
        return (round((100 * acc_annoy / len(idx)), 3), round(time_annoy, 2))
    elif all(model in model_type for model in ['faiss', 'annoy']):
        return (round((100 * acc_faiss / len(idx)), 3), round(time_faiss, 2),
                round((100 * acc_annoy / len(idx)), 3), round(time_annoy, 2))
```

```python
%%time

# Получим первичные значения целевой метрики accuracy@5 на базовых моделях.
(acc_faiss, time_faiss,
 acc_annoy, time_annoy)  = get_accuracy(
      df=data_base, df_eval=data_train, target=target_train,
      k_similar=5,  # Добавлено указание числа похожих объектов
      model_type=['faiss', 'annoy'],
      faiss_n_clust= 200,
      faiss_train_sample_size=int(data_base.shape[0] * .1),
      annoy_n_trees= 100
      )

# Создадим и выведем таблицу с результатами.
columns = pd.MultiIndex.from_tuples((('Модель FAISS', 'accuracy@5:'),
                                    ('Модель FAISS', 'Время поиска (сек.):'),
                                    ('Модель ANNOY', 'accuracy@5:'),
                                    ('Модель ANNOY', 'Время поиска (сек.):')))

base_results = pd.DataFrame([[acc_faiss, time_faiss, acc_annoy, time_annoy]],
                            columns=columns)

base_results
```

## Определение оптимального количества кластеров

```python
# Выбор подмножества данных из data_base_processed
subset_data = data_base_processed.sample(n=25000, random_state=42)

# Промежуток проверяемых значений количества кластеров
cluster_range = [5, 10, 25, 50, 100, 200, 400, 600, 1000]

# Функция для метода локтя (Elbow Method)
def elbow_method(data, cluster_range):
    inertias = []
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(data)
        inertias.append(kmeans.inertia_)
    plt.plot(cluster_range, inertias, marker='o')
    plt.xlabel('Количество кластеров')
    plt.ylabel('Inertia')
    plt.title('Метод локтя')
    plt.show()

# Функция для индекса Silhouette
def silhouette_index(data, cluster_range):
    silhouette_scores = []
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(data)
        score = silhouette_score(data, kmeans.labels_)
        silhouette_scores.append(score)
    plt.plot(cluster_range, silhouette_scores, marker='o')
    plt.xlabel('Количество кластеров')
    plt.ylabel('Silhouette Score')
    plt.title('Индекс Silhouette')
    plt.show()

# Применение метода локтя
elbow_method(subset_data, cluster_range)

# Применение индекса Silhouette
silhouette_index(subset_data, cluster_range)
```

Метод локтя не показывает явного излома (точки "локтя") и график индекса Silhouette почти постоянно растет с увеличением количества кластеров, но при этом достигает максимального значения около 0.07, не предоставляется возможным пока что определить оптимальное количество, попробуем воспользоваться еще следующей стратегией для выбора оптимального количества кластеров, такой как индекс Davies–Bouldin.

Вычислим индекс Davies–Bouldin для каждого количества кластеров в указанном промежутке и выведем результаты. Минимальное значение индекса Davies–Bouldin указывает на лучшую кластеризацию.

```python
# Выбор подмножества данных из data_base_processed
subset_data = data_base_processed.sample(n=25000, random_state=42)

# Промежуток проверяемых значений количества кластеров
cluster_range = [5, 10, 25, 50, 100, 200, 400, 600, 1000]

# Функция для индекса Davies–Bouldin
def davies_bouldin_index(data, cluster_range):
    db_scores = []
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(data)
        labels = kmeans.labels_
        db_score = davies_bouldin_score(data, labels)
        db_scores.append(db_score)
    return db_scores

# Вычисление индекса Davies–Bouldin для каждого количества кластеров
db_scores = davies_bouldin_index(subset_data, cluster_range)

# Вывод результатов
for k, db_score in zip(cluster_range, db_scores):
    print(f"Количество кластеров: {k}, Индекс Davies–Bouldin: {db_score}")
```

### Вывод

Также одной из причин уменьшения значения коэффициента Девиса-Болдина при увеличении количества кластеров может быть наличие большого оптимального числа кластеров. 

Однако из-за ограничений на технические возможности разделения данных на столь большое количество кластеров, мы ограничиваемся перебором количества кластеров в диапазоне от 100 до 500 и числа деревьев в диапазоне от 50 до 100 в циклах.
